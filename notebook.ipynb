```python
import os
import numpy as np
import joblib
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

def ensure_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)

def scaling(features, df, df_test=None, save_dir="./model"):
    ensure_dir(save_dir)

    df = df.copy()
    if df_test is not None:
        df_test = df_test.copy()

    for feature in features:
        scaler = MinMaxScaler()
        X_train = np.array(df[feature]).reshape(-1,1)
        scaler.fit(X_train)
        df[feature] = scaler.transform(X_train)
        joblib.dump(scaler, os.path.join(save_dir, f"scaler_{feature}.joblib"))

        if df_test is not None:
            X_test = np.array(df_test[feature]).reshape(-1,1)
            df_test[feature] = scaler.transform(X_test)

    if df_test is not None:
        return df, df_test
    else:
        return df

def encoding(features, df, df_test=None, save_dir="./model"):
    ensure_dir(save_dir)

    df = df.copy()
    if df_test is not None:
        df_test = df_test.copy()

    for feature in features:
        encoder = LabelEncoder()
        encoder.fit(df[feature])
        df[feature] = encoder.transform(df[feature])

        if df_test is not None:
            df_test[feature] = df_test[feature].apply(
                lambda x: x if x in encoder.classes_ else encoder.classes_[0]
            )
            df_test[feature] = encoder.transform(df_test[feature])

        joblib.dump(encoder, os.path.join(save_dir, f"encoder_{feature}.joblib"))

    if df_test is not None:
        return df, df_test
    else:
        return df
```

```python
def plot_feature_importances(feature_importances, cols, top_n=20, return_all=False):

    features = pd.DataFrame(feature_importances, columns=['coef_value'], index=cols)
    features = features.sort_values(by='coef_value', ascending=False)
    top_features = features.head(top_n)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='coef_value', y=top_features.index, data=top_features, palette='viridis')
    plt.title(f"Top {top_n} Feature Importances", fontsize=14)
    plt.xlabel("Mean Absolute Coefficient Value", fontsize=12)
    plt.ylabel("Features", fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

    return features if return_all else top_features
```

```python
def plot_feature_importances(importances, feature_names):
    import matplotlib.pyplot as plt
    import numpy as np

    indices = np.argsort(importances)[::-1]
    sorted_importances = importances[indices]
    sorted_features = [feature_names[i] for i in indices]

    plt.figure(figsize=(10, 6))
    plt.title('Feature Importances', fontsize=16)
    plt.barh(range(len(sorted_importances)), sorted_importances[::-1], color='skyblue')
    plt.yticks(range(len(sorted_features)), sorted_features[::-1], fontsize=12)
    plt.xlabel('Importance', fontsize=14)
    plt.tight_layout()
    plt.show()

plot_feature_importances(tree_model.feature_importances_, train_pca_df.columns)
```
